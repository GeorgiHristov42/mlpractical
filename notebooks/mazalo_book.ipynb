```python
def plot_single_stats(data, experiment_name = 'samlpe' ,notebook=True):
    
    plt.style.use('ggplot')
    fig_1 = plt.figure(figsize=(8, 4))
    ax_1 = fig_1.add_subplot(111)
    ax_1.plot(data['epoch'], data['val_c_accuracy'], color='r', label='Validation(acc)')
    ax_1.plot(data['epoch'], data['train_c_accuracy'], color='b', label='Train(acc)')
    ax_1.legend(loc=0)
    ax_1.set_xlabel('Epoch number')
    fig_1.savefig

    fig_2 = plt.figure(figsize=(8, 4))
    ax_2 = fig_2.add_subplot(111)
    ax_2.plot(data['epoch'], data['train_c_loss'], color='r', label='Validation(err)')
    ax_2.plot(data['epoch'], data['val_c_loss'], color='b', label='Train(err)')
    ax_2.legend(loc=0)
    ax_2.set_xlabel('Epoch number')

    fig_1.savefig('Experiment_' + str(experiment_name) + '_acc.pdf')
    fig_2.savefig('Experiment_' + str(experiment_name) + '_err.pdf')
```

```python
def plot_multiple_stats(data, experiment_name = 'samlpe' ,notebook=True):
    
    plt.style.use('ggplot')
    fig_1 = plt.figure(figsize=(8, 4))
    ax_1 = fig_1.add_subplot(111)
    for d,l in data:
        ax_1.plot(d['epoch'], d['val_c_accuracy'], label= str(l) + '(val)')
        ax_1.plot(d['epoch'], d['train_c_accuracy'], label=str(l) + '(train)')
    ax_1.legend(loc=0)
    ax_1.set_xlabel('Epoch number')
    fig_1.savefig

#     fig_2 = plt.figure(figsize=(8, 4))
#     ax_2 = fig_2.add_subplot(111)
#     ax_2.plot(data['epoch'], data['train_c_loss'], color='r', label='Val(err)')
#     ax_2.plot(data['epoch'], data['val_c_loss'], color='b', label='Train(err)')
#     ax_2.legend(loc=0)
#     ax_2.set_xlabel('Epoch number')

    fig_1.savefig('images/Experiment_' + str(experiment_name) + '_acc.pdf')
#     fig_2.savefig('Experiment_' + str(experiment_name) + '_err.pdf')
```
